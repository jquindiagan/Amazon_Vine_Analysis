```python
import pandas as pd
```

```python
# Read vine_table into dataframe
df = pd.read_csv ('vine_table.csv')
df.columns = ["review_id", "star_rating", "helpful_votes", "total_votes", "vine", "verified_purchase"]
df
```

```python
# 1. Filter the data and create a new DataFrame or table to retrieve all the rows where the total_votes count is equal to or
# greater than 20 to pick reviews that are more likely to be helpful and to avoid having division by zero errors later on.
filter_df = df[df.total_votes >= 20.0]
filter_df
```

```python
# 2. Filter the new DataFrame or table created in Step 1 and create a new DataFrame or table to retrieve all the rows
# where the number of helpful_votes divided by total_votes is equal to or greater than 50%.
filter2_df = filter_df[(filter_df.helpful_votes/filter_df.total_votes) >= 0.5]
filter2_df
```

```python
# 3. Filter the DataFrame or table created in Step 2, and create a new DataFrame or table that retrieves all the rows
# where a review was written as part of the Vine program (paid), vine == 'Y'.
filter3_df = filter2_df[filter2_df.vine == "Y"]
filter3_df
```

```python
# 4. Repeat Step 3, but this time retrieve all the rows where the review was not part of the Vine program (unpaid),
# vine == 'N'.
filter4_df = filter2_df[filter2_df.vine == "N"]
filter4_df
```

```python
# 5a. Determine the total number of reviews for the two types of review (paid vs unpaid).
# Total unpaid reviews
total_unpaid_reviews = df[df.vine == 'N']
total_unpaid_reviews_count = total_unpaid_reviews['review_id'].count()
total_unpaid_reviews_count
```

```python
# Total paid reviews
total_paid_reviews = df[df.vine == 'Y']
total_paid_reviews_count = total_paid_reviews['review_id'].count()
total_paid_reviews_count
```

```python
# 5b. Determine the number of 5-star reviews, and the percentage of 5-star reviews for the two types of review (paid vs unpaid).
# Unpaid 5-star reviews
unpaid_5_star = total_unpaid_reviews[total_unpaid_reviews.star_rating == 5]
unpaid_5_star_count = unpaid_5_star['review_id'].count()
unpaid_5_star_count
```

```python
# Paid 5-star reviews
paid_5_star = total_paid_reviews[total_paid_reviews.star_rating == 5]
paid_5_star_count = paid_5_star['review_id'].count()
paid_5_star_count
```

```python
# 5c. Determine the percentage of 5-star reviews for the two types of review (paid vs unpaid).
# Unpaid 5-star percentage
unpaid_5_per = (unpaid_5_star_count/total_unpaid_reviews_count) * 100
unpaid_5_per
```

```python
# 5c. Determine the percentage of 5-star reviews for the two types of review (paid vs unpaid).
# Paid 5-star percentage
paid_5_per = (paid_5_star_count/total_paid_reviews_count) * 100
paid_5_per
```
